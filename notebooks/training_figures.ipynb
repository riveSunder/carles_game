```python
# CornerBonus Training with Toggle

filenames = [\
            "Toggle_43110_default_tag1622338849_resultsdefault_tag1622338849.npy", \
            "Toggle_12345_default_tag1622316512_resultsdefault_tag1622316512.npy", \
            "Toggle_1337_default_tag1622274062_resultsdefault_tag1622274062.npy", \
            "Toggle_42_default_tag1622295184_resultsdefault_tag1622295184.npy", \
            "Toggle_13_default_tag1622252874_resultsdefault_tag1622252874.npy" \
            ]

plt.figure(figsize=(18,10))
for filename in filenames:
    
    agent_label = filename.split("_")[0]
    seed_label = filename.split("_")[1]
    
    my_data = np.load(f"../experiments/logs/{filename}", allow_pickle=True)[np.newaxis][0]
    
    plt.plot(my_data["generation"], my_data["fitness_mean"], lw=4, label=f"{agent_label} agent, seed: {seed_label}")

    my_data["fitness_mean"] = np.array(my_data["fitness_mean"]) #/ len(my_data["fitness_mean"])
    my_data["fitness std. dev."] = np.array(my_data["fitness std. dev."]) #/ len(my_data["fitness_mean"])

    plt.fill_between(my_data["generation"], my_data["fitness_mean"]-my_data["fitness std. dev."], my_data["fitness_mean"]+my_data["fitness std. dev."], alpha=0.125)
    
plt.plot(my_data["generation"], [0.0 for elem in my_data["generation"]],  '--',color = [0,0,0], alpha=0.5)
plt.legend()
plt.title("CornerBonus with Toggle Agent", fontsize = my_fontsize + 12)
plt.ylabel("Cumulative Reward (mean)", fontsize = my_fontsize + 6)
plt.xlabel("Generation", fontsize = my_fontsize + 6)
plt.axis((0, len(my_data["generation"]), -3e6, 1e6))
```
